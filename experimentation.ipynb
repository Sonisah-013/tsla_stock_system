```python
import os, joblib, numpy as np, nest_asyncio, uvicorn
from fastapi import FastAPI, Request
from fastapi.responses import HTMLResponse
from fastapi.templating import Jinja2Templates
from pydantic import BaseModel

nest_asyncio.apply()
app = FastAPI()

# FORCE THE PATHS - This looks exactly where your files are
BASE_DIR = os.getcwd()
# If you are inside the 'notebooks' folder, go up one level
if "notebooks" in BASE_DIR:
    BASE_DIR = os.path.dirname(BASE_DIR)

MODEL_PATH = os.path.join(BASE_DIR, "saved_models", "tsla_model.pkl")
TEMPLATES_PATH = os.path.join(BASE_DIR, "dashboard", "templates")

model = joblib.load(MODEL_PATH)
templates = Jinja2Templates(directory=TEMPLATES_PATH)

class StockInput(BaseModel):
    open: float; high: float; low: float; volume: float

@app.get("/", response_class=HTMLResponse)
async def home(request: Request):
    return templates.TemplateResponse("index.html", {"request": request})

@app.post("/predict")
async def predict(data: StockInput):
    features = np.array([[data.open, data.high, data.low, data.volume]])
    prediction = model.predict(features)[0]
    return {"status": "success", "predicted_close": round(float(prediction), 2)}

if __name__ == "__main__":
    print(f"🚀 SERVER STARTING ON PORT 8081")
    print(f"🔗 CLICK HERE: http://127.0.0.1:8081")
    # CHANGED PORT TO 8081 TO AVOID CONFLICTS
    config = uvicorn.Config(app, host="127.0.0.1", port=8081, loop="asyncio")
    server = uvicorn.Server(config)
    await server.serve()
```
